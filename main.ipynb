```python
import torch
import torchvision
import torchvision.transforms as transforms
import numpy as np
import os
import matplotlib.pyplot as plt

import npy_loader_src



batch_size = 4

# Enter the absolute path to the dataset folder

val_dir = os.path.join(os.getcwd(), 'dataset', 'val')
train_dir = os.path.join(os.getcwd(), 'dataset', 'train')

# Using custom loader to load .npy files

trainset = torchvision.datasets.DatasetFolder(root=train_dir, loader=npy_loader_src.npy_loader, extensions='.npy')
trainloader = torch.utils.data.DataLoader(trainset, batch_size=batch_size,
                                          shuffle=True, num_workers=2)

valset = torchvision.datasets.DatasetFolder(root=val_dir, loader=npy_loader_src.npy_loader, extensions='.npy')
valloader = torch.utils.data.DataLoader(valset, batch_size=batch_size,
                                        shuffle=False, num_workers=2)

```

```python

classes = ['no', 'sphere', 'vort']

# functions to show an image


def imshow(img): # img is a tensor
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()


# get some random training images
dataiter = iter(trainloader)
images, labels = next(dataiter)

# show images
imshow(torchvision.utils.make_grid(images))
# print labels
print(' '.join(f'{classes[labels[j]]:5s}' for j in range(batch_size)))
```
